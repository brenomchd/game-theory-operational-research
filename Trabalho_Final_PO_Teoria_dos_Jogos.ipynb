## Problemas Modelados através de Jogos

### Definição de Estratégias de Marketing entre Duas Empresas utilizando Teoria dos Jogos e Programação Linear



# 1. Introdução

Empresas competem constantemente pela atenção dos consumidores. Neste trabalho, modelamos essa competição por meio da Teoria dos Jogos.

Duas empresas disputam participação de mercado utilizando três estratégias:

- TV
- Redes Sociais
- Influenciadores

O problema será tratado como um jogo de soma zero e resolvido por Programação Linear.



# 2. Conceitos de Teoria dos Jogos

- **Jogadores:** Empresa A e Empresa B.
- **Estratégias:** ações disponíveis para cada jogador.
- **Payoff:** ganho ou vantagem competitiva obtida.
- **Jogo de Soma Zero:** o ganho de uma empresa representa a perda relativa da outra.
- **Estratégias Mistas:** utilização probabilística das estratégias disponíveis.



# 3. Modelagem do Problema

Duas empresas concorrentes disputam o mesmo mercado.

Cada empresa deve escolher uma das seguintes estratégias:

1. TV
2. Redes Sociais
3. Influenciadores

A interação entre as empresas será representada por uma matriz de payoff.


```

import numpy as np
import pandas as pd

estrategias = ["TV", "Redes Sociais", "Influenciadores"]

payoff_base = np.array([
    [1, -2, 3],
    [2, 1, -1],
    [-3, 2, 0]
])

pd.DataFrame(
    payoff_base,
    index=estrategias,
    columns=estrategias
)
```


# 4. Formulação em Programação Linear

O objetivo é encontrar a estratégia ótima para a Empresa A.

Restrições:

- As probabilidades devem ser não negativas.
- A soma das probabilidades deve ser igual a 1.
- O ganho esperado deve ser maximizado.

O problema será resolvido utilizando o método Simplex através do solver `linprog`.


```

from scipy.optimize import linprog
```


# 5. Função de Resolução do Jogo


```

def resolver_jogo(payoff):

    K = abs(payoff.min()) + 1

    A = payoff + K

    m, n = A.shape

    resultado = linprog(
        c=np.ones(m),
        A_ub=-A.T,
        b_ub=-np.ones(n),
        bounds=[(0, None)] * m,
        method="highs"
    )

    y = resultado.x

    soma = y.sum()

    probabilidades = y / soma

    valor_jogo = 1 / soma - K

    return probabilidades, valor_jogo
```


# 6. Resultados do Cenário Base


```

probs, valor = resolver_jogo(payoff_base)

resultado_df = pd.DataFrame({
    "Estratégia": estrategias,
    "Probabilidade Ótima": probs
})

resultado_df
```

```

print("Estratégia ótima da Empresa A\n")

for e, p in zip(estrategias, probs):
    print(f"{e}: {p:.4f}")

print(f"\nValor do jogo: {valor:.4f}")
```

```

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(estrategias, probs)
plt.title("Estratégia Ótima - Cenário Base")
plt.ylabel("Probabilidade")
plt.show()
```


# 7. Análise de Cenários

Serão avaliados três cenários:

- Cenário Base
- Redes Sociais mais eficientes
- Influenciadores mais eficientes


```

payoff_social = np.array([
    [1, -3, 3],
    [4, 2, 2],
    [-3, 1, 0]
])

payoff_influencer = np.array([
    [1, -2, 1],
    [2, 1, -2],
    [-1, 4, 3]
])

cenarios = {
    "Base": payoff_base,
    "Redes Sociais": payoff_social,
    "Influenciadores": payoff_influencer
}
```

```

dados = []

for nome, matriz in cenarios.items():

    probs, valor = resolver_jogo(matriz)

    print(f"\n=== {nome} ===")

    for e, p in zip(estrategias, probs):
        print(f"{e}: {p:.4f}")

    print(f"Valor do jogo: {valor:.4f}")

    for e, p in zip(estrategias, probs):
        dados.append([nome, e, p])
```

```

df = pd.DataFrame(
    dados,
    columns=["Cenário", "Estratégia", "Probabilidade"]
)

pivot = df.pivot(
    index="Estratégia",
    columns="Cenário",
    values="Probabilidade"
)

pivot
```

```

pivot.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Comparação das Estratégias Ótimas por Cenário")
plt.ylabel("Probabilidade")
plt.show()
```


# 8. Conclusão

A Teoria dos Jogos permitiu modelar a competição estratégica entre duas empresas.

Por meio da Programação Linear foi possível encontrar estratégias ótimas e analisar como mudanças no mercado afetam as decisões mais vantajosas para cada empresa.
